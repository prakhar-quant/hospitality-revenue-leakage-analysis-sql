```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
cities = ['Mumbai','Bangalore','Hyderabad','Delhi']
revenue = [668640991,420397050,325232870,294500318]

df = pd.DataFrame({
'City': cities,
'Revenue': revenue
})

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='City',y='Revenue')

plt.title('Revenue Generated by City',fontsize=16)
plt.ylabel('Revenue (₹)')
plt.xlabel('')

for i,v in enumerate(df['Revenue']):
    plt.text(i,v,f'₹{v/1e6:.1f}M',ha='center')

plt.show()
```

```python
city = ['Mumbai','Bangalore','Hyderabad','Delhi']

revenue = [668640991,420397050,325232870,294500318]

occupancy = [57.88,55.77,58.07,60.55]

plt.figure(figsize=(10,6))

plt.scatter(
    occupancy,
    revenue,
    s=np.array(revenue)/500000,
    alpha=0.7
)

for i in range(len(city)):
    plt.annotate(city,
                 (occupancy[i],revenue[i]))

plt.xlabel('Occupancy Rate (%)')
plt.ylabel('Revenue')
plt.title('Occupancy vs Revenue')

plt.show()
```
